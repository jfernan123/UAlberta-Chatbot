### Q20: What is the difference between the Mathematics and Statistics programs?

**Time:** 10.9s

---

- **Mathematics** emphasizes mathematical modeling, optimization, stochastic processes, and theoretical foundations of data science.  
- **Statistics** focuses on statistical modeling, inference, regression, and statistical learning methods for analyzing real-world data.
