# The Decima Robinson Support Centre

The **Decima Robinson Support Centre** is a comprehensive support service for students taking **100- and 200-level mathematics and statistics courses** at the University of Alberta.

## Location & Hours
- **Location:** University Commons 1-200
- **Hours:** 
  - Monday, Wednesday, Friday: 9 AM to 3 PM
  - Tuesday, Thursday: 10 AM to 4 PM
  - (Open during Fall and Winter terms when classes are taught)

## Services Offered
The Centre provides:
- **One-on-one help** with homework on a drop-in basis
- **Exam study sessions** for select courses
- **Precalculus program** for students refreshing high school math skills while taking calculus
- **Weekly review seminars**
- **Primer Courses** to help high school students transition to university

The Centre is staffed by graduate students who are available to assist you with your coursework.

## Contact Information
For questions about services or feedback, you can:
- **Visit:** University Commons 1-200
- **Email:** mssdrsc@ualberta.ca

## Historical Note
The Centre was named after **Decima Robinson**, the first BSc graduate at the University of Alberta, who graduated with a major in mathematics in 1911.
